```python
#import all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('USvideos.csv')
```

```python
genre=pd.read_json('US_category_id (1).json')
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['tags','thumbnail_link','description'],axis=1,inplace=True)
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'],format="%y.%d.%m")
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

```python
df['category_id']
```

```python
genre['items'][0]['id']
```

```python
genre['items'][0]['snippet']['title']
```

```python
id=[genre['items'][i]['id'] for i in range(len(genre))]
```

```python
gen=[genre['items'][i]['snippet']['title'] for i in range(len(genre))]
```

```python
data=pd.DataFrame({'category_id':id ,'genre':gen})
data
```

```python
data['category_id'] = data['category_id'].astype(int)
```

```python
finaldf = pd.merge(df,data,on = 'category_id',how='inner')
finaldf
```

```python
finaldf.to_csv('youtubecleaned.csv')
```
